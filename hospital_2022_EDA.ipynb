```python
# создаю df для последующей корреляции
df_corr = df.drop(columns=['Дата_поступления',
                           'Отделение_нахождения',
                           'Профильное_отделение',
                           'Направившее_учреждение',
                           'Дата_выписки',
                           'Результат_лечения',
                           'Код_диагноза_выписки_МКБ',
                           'Название_диагноза_выписки',
                           'Страна',
                           'Куда_выбыл',
                           'Тип_госпитализации'])
df_corr.head()
```

```python
df_corr = pd.get_dummies(df_corr, columns=['Пол','Исход_лечения', 'Вид_госпитализации'])
df_corr.head()
```

```python
df_corr = df_corr.drop(columns=['Исход_лечения_Без изменений',
                                'Исход_лечения_Выздоровление',
                                'Исход_лечения_С ухудшением',
                                'Вид_госпитализации_Неотложка'])
df_corr.head()
```

```python
# использую корреляцию Спирмена
df_corr.corr('spearman')
```

```python
correlation = df_corr.corrwith(df_corr['Исход_лечения_Смерть'])
print(correlation)
```

```python
plt.figure(figsize=(10, 6))
heatmap = sns.heatmap(df_corr.corr(), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Коэффициенты корреляции')
plt.show()
```

```python
df_corr['Средняя_смертность_кд'] = df_corr.groupby(['Количество_койко_дней'])['Исход_лечения_Смерть'].transform('mean')
df_corr
```

```python
plt.figure(figsize=(8, 4))
plt.scatter(df_corr['Количество_койко_дней'], df_corr['Средняя_смертность_кд'], alpha=0.5, s=4)
plt.title('Зависимость между смертностью и продолжительностью госпитализации')
plt.xlabel('Продолжительность госпитализации')
plt.ylabel('Смертность')
plt.show()
```
